# Scrape the Data from wikepedia and create a pandas dataframe 


## scrape the data from the given URL 

```python
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
import pandas as pd
```

```python
res=response.content
```

## Transform data into pandas dataframe

```python
df_list = pd.read_html(res)
```

```python
df = df_list[0]
```

```python
print(df)
```

## Remove 'Not assigned' from Borough Column

```python
df = df[(df.Borough!='Not assigned')]
```
